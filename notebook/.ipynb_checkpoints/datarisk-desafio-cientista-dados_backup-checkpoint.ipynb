### Resumo parcial

- A análise do tipo de variável permite checar se foram atribuídos corretamente os tipos de dados e quando constatado alguma inconsistência fazer as devidas correções para evitar problemas em análises futuras, como por exemplo o valor **int** usado para `id_cliente` e `id_contrato`

Quanto aos valores ausentes, pode significar falha no preenchimento, mas também pode carregar informação de não pagamento, que é o que buscamos.

```python
# Altera o tipo de variável

df_cadastro["id_cliente"] = df_cadastro["id_cliente"].astype("object") 
df_cadastro["data_nascimento"] = pd.to_datetime(df_cadastro["data_nascimento"], errors="coerce")
df_cadastro["idade"] = (datetime.today().year - df_cadastro["data_nascimento"].dt.year)

df_emprestimo["id_cliente"] = df_emprestimo["id_cliente"].astype("object")
df_emprestimo["id_contrato"] = df_emprestimo["id_contrato"].astype("object")

colunas_datas_emprestimo = [
    "data_decisao",
    "data_encerramento",
    "data_liberacao",
    "data_primeiro_vencimento",
    "data_ultimo_vencimento",
    "data_ultimo_vencimento_original"]

for col in colunas_datas_emprestimo:
    if col in df_emprestimo.columns:
        df_emprestimo[col] = pd.to_datetime(df_emprestimo[col], errors="coerce")



df_parcelas["id_cliente"] = df_parcelas["id_cliente"].astype("object")
df_parcelas["id_contrato"] = df_parcelas["id_contrato"].astype("object")
df_parcelas["data_prevista_pagamento"] = pd.to_datetime(df_parcelas["data_prevista_pagamento"])
df_parcelas["data_real_pagamento"]= pd.to_datetime(df_parcelas["data_real_pagamento"])

df_parcelas = df_parcelas.groupby(["id_cliente","id_contrato", "numero_parcela"]).agg({
    "data_prevista_pagamento":"max",
    "data_real_pagamento":"max",
    "valor_previsto_parcela":"first",
    "valor_pago_parcela":"sum"
}).reset_index()

df_parcelas["dias_atraso"] = (df_parcelas['data_real_pagamento'] - df_parcelas['data_prevista_pagamento']).dt.days
# df_parcelas['dias_atraso'] = df_parcelas['dias_atraso'].fillna(0)
```

No contexto de **risco de crédito**, existe diversos segmentos e para cada um deles é aplicado um modelo diferente com regras de negócio para aquele tipo de segmento.

- **Cash Loans:** Emprestimo pessoal
  
- **Revolving Loans:** Crédito rotativo.

  
- **Consumer Loans:** Financiamento

  
- XNA indica ausência de informação


A decisão será escolher o segmento com maior volume e modelar somente para esse **público alvo**, para evitar ruído de públicos cruzados com comportamentos distintos que pode interferir na capacidade preditiva do modelo.

O segmento escolhido é **Cash Loans**

```python
%%time
# Função para calcular as features ever30mob para um dado número de meses
def calculate_ever30mob_for_contract(df_parcelas, months):
    # Filtra as parcelas dentro do número especificado de meses para cada contrato
    df_window = df_parcelas[df_parcelas['numero_parcela'] <= months]

    # Agrupa por id_contrato e verifica se alguma parcela dentro da janela teve atraso de 30 ou mais dias
    ever30_flag = df_window.groupby('id_contrato')['dias_atraso'].apply(lambda x: ((x > 30) | (x.isna())).any()).astype(int)

    return ever30_flag.rename(f'ever30mob{str(months).zfill(2)}')

# Calcula para 3, 6 e 12 meses
ever30mob03_series = calculate_ever30mob_for_contract(df_parcelas, 3)
ever30mob06_series = calculate_ever30mob_for_contract(df_parcelas, 6)
ever30mob12_series = calculate_ever30mob_for_contract(df_parcelas, 12)

# Combina em um único DataFrame de features em nível de contrato
df_contract_mob_features = pd.DataFrame({
    'ever30mob03': ever30mob03_series,
    'ever30mob06': ever30mob06_series,
    'ever30mob12': ever30mob12_series
}).reset_index()

# Identifica as colunas 'ever30mob' que já existem em df_emprestimo
existing_ever30mob_cols = [col for col in df_emprestimo.columns if col.startswith('ever30mob')]

# Remove as colunas 'ever30mob' existentes de df_emprestimo antes do merge
# para evitar a criação de sufixos (_x, _y) e garantir que as novas sejam usadas.
if existing_ever30mob_cols:
    df_emprestimo = df_emprestimo.drop(columns=existing_ever30mob_cols)

# Mescla estas novas features no df_emprestimo
print(df_emprestimo.shape)
df_emprestimo = df_emprestimo.merge(df_contract_mob_features, on='id_contrato', how='left')
print(df_emprestimo.shape)

# Preenche valores NaN que podem surgir se um contrato em df_emprestimo não tiver parcelas correspondentes
# ou não tiver parcelas dentro da janela, indicando que não houve atraso de 30+ dias. O padrão é 0.
df_emprestimo['ever30mob03'] = df_emprestimo['ever30mob03'].fillna(0).astype(int)
df_emprestimo['ever30mob06'] = df_emprestimo['ever30mob06'].fillna(0).astype(int)
df_emprestimo['ever30mob12'] = df_emprestimo['ever30mob12'].fillna(0).astype(int)
```

```python
df_emprestimo['data_decisao_mes_ano'] = df_emprestimo['data_decisao'].dt.to_period('M')

defaults_over_time = df_emprestimo.groupby('data_decisao_mes_ano')[['ever30mob03', 'ever30mob06', 'ever30mob12']].sum().reset_index()

defaults_over_time['data_decisao_mes_ano'] = defaults_over_time['data_decisao_mes_ano'].astype(str)

plt.figure(figsize=(15, 7))

sns.lineplot(x='data_decisao_mes_ano', y='ever30mob03', data=defaults_over_time, label='Atraso de 30+ dias nos primeiros 3 meses')
sns.lineplot(x='data_decisao_mes_ano', y='ever30mob06', data=defaults_over_time, label='Atraso de 30+ dias nos primeiros 6 meses')
sns.lineplot(x='data_decisao_mes_ano', y='ever30mob12', data=defaults_over_time, label='Atraso de 30+ dias nos primeiros 12 meses')

plt.title('Quantidade de Atrasos ao Longo do Tempo')
plt.xlabel('Data (Mês-Ano)')
plt.ylabel('Número de Atrasos')
plt.xticks(rotation=90, ha='right')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
